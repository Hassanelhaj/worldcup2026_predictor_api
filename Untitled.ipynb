```python
# ╔══════════════════════════════════════════════════════════════════╗
# ║  CELL 1 — Install dependencies (run once)                       ║
# ╚══════════════════════════════════════════════════════════════════╝

# !pip install pandas numpy scikit-learn xgboost joblib requests scipy --quiet


# ╔══════════════════════════════════════════════════════════════════╗
# ║  CELL 2 — Full pipeline: train models + save all artifacts      ║
# ║  Just run this cell — it does everything automatically          ║
# ╚══════════════════════════════════════════════════════════════════╝

import os, warnings, urllib.request
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import joblib

from collections import defaultdict
from scipy.stats import poisson as scipy_poisson
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import PoissonRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, log_loss
import xgboost as xgb

np.random.seed(42)

# ── Output folder: models/ next to this notebook ──────────────────
MODELS_DIR = os.path.join(os.getcwd(), "models")
os.makedirs(MODELS_DIR, exist_ok=True)
print(f"📁 Models will be saved to: {MODELS_DIR}")

# ══════════════════════════════════════════════════════════════════
# STEP 1 — Download data
# ══════════════════════════════════════════════════════════════════
DATA_PATH = "results.csv"
if not os.path.exists(DATA_PATH):
    print("⬇️  Downloading match data...")
    urllib.request.urlretrieve(
        "https://raw.githubusercontent.com/martj42/international_results/master/results.csv",
        DATA_PATH
    )
print(f"✅ Data ready: {DATA_PATH}")

# ══════════════════════════════════════════════════════════════════
# STEP 2 — Load & clean
# ══════════════════════════════════════════════════════════════════
EXTRA_NAMES = {
    "German DR":"Germany","East Germany":"Germany","West Germany":"Germany",
    "USSR":"Russia","Soviet Union":"Russia","CIS":"Russia",
    "Czechoslovakia":"Czech Republic",
    "Yugoslavia":"Serbia","FR Yugoslavia":"Serbia","Serbia and Montenegro":"Serbia",
    "Netherlands Antilles":"Curaçao","Swaziland":"Eswatini",
    "Macedonia":"North Macedonia",
    "Irish Free State":"Republic of Ireland","Éire":"Republic of Ireland",
    "Zaire":"DR Congo","Zaïre":"DR Congo","Western Samoa":"Samoa","Burma":"Myanmar",
}

# Load former_names.csv if it exists, otherwise just use EXTRA_NAMES
name_map = dict(EXTRA_NAMES)
if os.path.exists("former_names.csv"):
    fn = pd.read_csv("former_names.csv")
    for _, r in fn.iterrows():
        name_map[r["former"]] = r["current"]
    print(f"✅ Loaded {len(name_map)} name mappings from former_names.csv")
else:
    print(f"ℹ️  former_names.csv not found — using {len(name_map)} built-in mappings")

df = pd.read_csv(DATA_PATH, parse_dates=["date"])
df["home_score"] = pd.to_numeric(df["home_score"], errors="coerce")
df["away_score"] = pd.to_numeric(df["away_score"], errors="coerce")
df.dropna(subset=["home_score","away_score"], inplace=True)
df["home_score"] = df["home_score"].astype(int)
df["away_score"] = df["away_score"].astype(int)
df["home_team"]  = df["home_team"].replace(name_map)
df["away_team"]  = df["away_team"].replace(name_map)
df.sort_values("date", inplace=True)
df.reset_index(drop=True, inplace=True)
print(f"✅ {len(df):,} matches loaded ({df['date'].min().year}–{df['date'].max().year})")

# ══════════════════════════════════════════════════════════════════
# STEP 3 — ELO ratings
# ══════════════════════════════════════════════════════════════════
INITIAL_ELO = 1500; K = 32; HOME_ADV = 100

elo = defaultdict(lambda: INITIAL_ELO)
heb, aeb = [], []

for _, row in df.iterrows():
    h, a    = row["home_team"], row["away_team"]
    neutral = bool(row["neutral"])
    eh = elo[h] + (0 if neutral else HOME_ADV)
    ea = elo[a]
    exp_h = 1.0 / (1.0 + 10**((ea - eh) / 400.0))
    sh, sa = int(row["home_score"]), int(row["away_score"])
    if sh > sa:    sc_h, sc_a = 1.0, 0.0
    elif sh == sa: sc_h, sc_a = 0.5, 0.5
    else:          sc_h, sc_a = 0.0, 1.0
    heb.append(elo[h]); aeb.append(elo[a])
    elo[h] += K * (sc_h - exp_h)
    elo[a] += K * (sc_a - (1.0 - exp_h))

df["home_elo_before"] = heb
df["away_elo_before"] = aeb
elo_ratings = dict(elo)

top5 = sorted(elo.items(), key=lambda x: -x[1])[:5]
print("✅ ELO computed. Top 5:", " | ".join(f"{k} {v:.0f}" for k,v in top5))

# ══════════════════════════════════════════════════════════════════
# STEP 4 — Feature engineering (vectorised)
# ══════════════════════════════════════════════════════════════════
CONF_STR = {"UEFA":1.08,"CONMEBOL":1.06,"CONCACAF":0.96,"CAF":0.94,"AFC":0.93,"OFC":0.88}
CONF_MAP = {
    **{t:"UEFA" for t in ["Germany","France","Spain","Italy","England","Portugal",
       "Netherlands","Belgium","Croatia","Switzerland","Poland","Denmark","Sweden",
       "Austria","Serbia","Ukraine","Czech Republic","Hungary","Slovakia","Romania",
       "Turkey","Greece","Scotland","Norway","Finland","Wales","Slovenia",
       "North Macedonia","Albania","Kosovo","Montenegro","Iceland",
       "Republic of Ireland","Northern Ireland","Belarus","Bulgaria","Russia","Georgia"]},
    **{t:"CONMEBOL" for t in ["Brazil","Argentina","Uruguay","Colombia","Chile",
       "Ecuador","Peru","Paraguay","Bolivia","Venezuela"]},
    **{t:"CONCACAF" for t in ["United States","Mexico","Canada","Costa Rica","Honduras",
       "Jamaica","Panama","El Salvador","Trinidad and Tobago","Haiti","Guatemala",
       "Cuba","Curaçao","Suriname","Guyana","Belize","Nicaragua"]},
    **{t:"CAF" for t in ["Senegal","Morocco","Egypt","Nigeria","Cameroon","Tunisia",
       "Ghana","Algeria","Ivory Coast","Mali","Burkina Faso","Guinea","Congo",
       "DR Congo","Zimbabwe","Zambia","Uganda","Tanzania","Kenya","Ethiopia",
       "Cape Verde","Benin","Gabon","Rwanda","Mozambique","Madagascar",
       "Guinea-Bissau","Equatorial Guinea","Mauritania","Sudan","South Africa",
       "Angola","Togo","Niger","Sierra Leone","Liberia","Djibouti","Eswatini",
       "Malawi","Namibia","Botswana","Lesotho"]},
    **{t:"AFC" for t in ["Japan","South Korea","Iran","Australia","Saudi Arabia","Qatar",
       "UAE","Iraq","China","Uzbekistan","Jordan","Oman","Bahrain","Kuwait","Vietnam",
       "Thailand","India","Indonesia","Malaysia","Philippines","Myanmar","North Korea",
       "Syria","Lebanon","Palestine","Yemen","Tajikistan","Kyrgyzstan","Nepal",
       "Sri Lanka","Bangladesh","Pakistan","Afghanistan"]},
    **{t:"OFC" for t in ["New Zealand","Fiji","Papua New Guinea","Vanuatu",
       "Solomon Islands","New Caledonia","Tahiti","Samoa","Tonga","Cook Islands"]},
}

def get_conf_str(team):
    return CONF_STR.get(CONF_MAP.get(team, ""), 1.0)

df2 = df[df["date"].dt.year >= 1990].copy().reset_index(drop=True)

hr = df2[["date","home_team","home_score","away_score"]].copy()
hr.columns = ["date","team","scored","conceded"]
ar = df2[["date","away_team","away_score","home_score"]].copy()
ar.columns = ["date","team","scored","conceded"]
long = pd.concat([hr, ar], ignore_index=True)
long["won"] = np.where(long.scored > long.conceded, 1.0,
              np.where(long.scored == long.conceded, 0.5, 0.0))
long["gd"]  = long.scored - long.conceded
long.sort_values(["team","date"], inplace=True)

long["form5"] = long.groupby("team")["won"].transform(
    lambda s: s.ewm(span=5, min_periods=1).mean().shift(1).fillna(0.5))
long["gd10"]  = long.groupby("team")["gd"].transform(
    lambda s: s.rolling(10, min_periods=1).mean().shift(1).fillna(0.0))
stats = long[["date","team","form5","gd10"]].drop_duplicates(["team","date"])

feat = df2.merge(
    stats.rename(columns={"team":"home_team","form5":"home_form_avg","gd10":"home_gd_last10"}),
    on=["date","home_team"], how="left")
feat = feat.merge(
    stats.rename(columns={"team":"away_team","form5":"away_form_avg","gd10":"away_gd_last10"}),
    on=["date","away_team"], how="left")

feat["home_form_avg"]   = feat["home_form_avg"].fillna(0.5)
feat["away_form_avg"]   = feat["away_form_avg"].fillna(0.5)
feat["home_gd_last10"]  = feat["home_gd_last10"].fillna(0.0)
feat["away_gd_last10"]  = feat["away_gd_last10"].fillna(0.0)
feat["elo_diff"]        = (feat["home_elo_before"] - feat["away_elo_before"]
                           + feat["neutral"].apply(lambda x: 0 if x else HOME_ADV))
feat["rest_diff"]       = 0.0   # no rest data in raw CSV; set to 0
feat["home_conf_mult"]  = feat["home_team"].map(get_conf_str)
feat["away_conf_mult"]  = feat["away_team"].map(get_conf_str)
feat["neutral"]         = feat["neutral"].astype(int)
feat["outcome"]         = np.where(feat["home_score"] > feat["away_score"], 2,
                          np.where(feat["home_score"] == feat["away_score"], 1, 0))

print(f"✅ Features built: {feat.shape}")

# ══════════════════════════════════════════════════════════════════
# STEP 5 — Train / validation split (WC matches only)
# ══════════════════════════════════════════════════════════════════
FEATURE_COLS = [
    "elo_diff", "rest_diff", "home_form_avg", "away_form_avg",
    "home_gd_last10", "away_gd_last10", "home_conf_mult", "away_conf_mult"
]

is_wc = feat["tournament"].str.contains("FIFA World Cup", na=False, case=False)
is_q  = feat["tournament"].str.contains("World Cup qualification", na=False, case=False)
wc_df = feat[is_wc & (feat["date"].dt.year >= 1998)]
q_df  = feat[is_q  & (feat["date"].dt.year >= 2006)]
wc22_ko = wc_df[(wc_df["date"].dt.year == 2022) & (wc_df["date"] >= "2022-12-03")]
train = pd.concat([wc_df[~wc_df.index.isin(wc22_ko.index)], q_df]).drop_duplicates()
val   = wc22_ko.copy()

X_tr  = train[FEATURE_COLS].fillna(0)
y_tr  = train["outcome"]
X_val = val[FEATURE_COLS].fillna(0)
y_val = val["outcome"]

print(f"✅ Train: {len(train):,}  |  Val: {len(val):,} (WC2022 knockouts)")

# ══════════════════════════════════════════════════════════════════
# STEP 6 — Train models
# ══════════════════════════════════════════════════════════════════
print("\n🏋️  Training models...")

# — Poisson (goal prediction) —
P_FEATS = ["elo_diff","home_form_avg","away_form_avg","neutral" if "neutral" in FEATURE_COLS
           else "elo_diff","home_gd_last10","away_gd_last10","home_conf_mult","away_conf_mult"]
P_FEATS = ["elo_diff","home_form_avg","away_form_avg",
           "home_gd_last10","away_gd_last10","home_conf_mult","away_conf_mult"]

poisson_home = PoissonRegressor(alpha=0.1, max_iter=300)
poisson_away = PoissonRegressor(alpha=0.1, max_iter=300)
poisson_home.fit(X_tr[P_FEATS], train["home_score"].clip(0, 10))
poisson_away.fit(X_tr[P_FEATS], train["away_score"].clip(0, 10))
print("  ✅ Poisson models trained")

# — Random Forest —
rf = RandomForestClassifier(
    n_estimators=300, max_depth=8, min_samples_leaf=5,
    random_state=42, n_jobs=-1
)
rf.fit(X_tr, y_tr)
rf_acc = accuracy_score(y_val, rf.predict(X_val))
print(f"  ✅ Random Forest trained  |  val accuracy = {rf_acc:.3f}")

# — XGBoost —
xgb_model = xgb.XGBClassifier(
    n_estimators=300, learning_rate=0.05, max_depth=5,
    subsample=0.8, colsample_bytree=0.8,
    objective="multi:softprob", num_class=3,
    eval_metric="mlogloss", random_state=42, verbosity=0
)
xgb_model.fit(X_tr, y_tr)
xgb_acc = accuracy_score(y_val, xgb_model.predict(X_val))
print(f"  ✅ XGBoost trained        |  val accuracy = {xgb_acc:.3f}")

# Pick best classifier
best_classifier = xgb_model if xgb_acc >= rf_acc else rf
best_name = "XGBoost" if xgb_acc >= rf_acc else "Random Forest"
print(f"\n🏆 Best classifier: {best_name} (acc={max(xgb_acc, rf_acc):.3f})")

# Label encoder (for team names — kept for API compatibility)
all_teams_list = sorted(set(feat["home_team"].tolist() + feat["away_team"].tolist()))
label_encoder = LabelEncoder()
label_encoder.fit(all_teams_list)
print(f"  ✅ Label encoder fitted on {len(all_teams_list)} teams")

# ══════════════════════════════════════════════════════════════════
# STEP 7 — Save all artifacts to models/
# ══════════════════════════════════════════════════════════════════
print(f"\n💾 Saving artifacts to {MODELS_DIR}/")

joblib.dump(best_classifier, os.path.join(MODELS_DIR, "best_classifier.pkl"))
print("  ✅ best_classifier.pkl")

joblib.dump(poisson_home,    os.path.join(MODELS_DIR, "poisson_home.pkl"))
print("  ✅ poisson_home.pkl")

joblib.dump(poisson_away,    os.path.join(MODELS_DIR, "poisson_away.pkl"))
print("  ✅ poisson_away.pkl")

joblib.dump(label_encoder,   os.path.join(MODELS_DIR, "label_encoder.pkl"))
print("  ✅ label_encoder.pkl")

joblib.dump(FEATURE_COLS,    os.path.join(MODELS_DIR, "feature_cols.pkl"))
print("  ✅ feature_cols.pkl")

# ══════════════════════════════════════════════════════════════════
# STEP 8 — Save / copy wc2026_predictions.csv to models/
# ══════════════════════════════════════════════════════════════════
CSV_SRC = "wc2026_predictions.csv"
CSV_DST = os.path.join(MODELS_DIR, "wc2026_predictions.csv")

if os.path.exists(CSV_SRC):
    import shutil
    shutil.copy(CSV_SRC, CSV_DST)
    print("  ✅ wc2026_predictions.csv  (copied from working directory)")
else:
    # Generate a fresh CSV from ELO-based estimates
    print("  ⚠️  wc2026_predictions.csv not found — generating from ELO estimates...")

    WC2026_TEAMS = [
        "Argentina","France","Brazil","Chile","Morocco","Uruguay","Peru","Canada","Bolivia",
        "England","Spain","Germany","Senegal","Algeria","Japan","Mexico","United States","Costa Rica",
        "Portugal","Netherlands","Colombia","Egypt","South Korea","Australia","Panama","Jamaica","New Zealand",
        "Belgium","Croatia","Switzerland","Nigeria","Cameroon","Ecuador","Honduras","Iran","Venezuela",
        "Italy","Denmark","Poland","Ghana","Saudi Arabia","Tunisia","Paraguay","El Salvador","Guatemala",
        "Serbia","DR Congo","Trinidad and Tobago",
    ]
    rows = []
    for team in WC2026_TEAMS:
        elo_val = elo_ratings.get(team, 1500)
        # Rough probability estimates based on ELO percentile
        base = max(0.1, (elo_val - 1400) / 1000 * 15)
        rows.append({
            "Team": team,
            "Round of 32":   f"{min(70, base*6):.1f}%",
            "Round of 16":   f"{min(50, base*4):.1f}%",
            "Quarter-final": f"{min(30, base*2.5):.1f}%",
            "Semi-final":    f"{min(20, base*1.5):.1f}%",
            "Final":         f"{min(12, base*0.9):.1f}%",
            "Champion":      f"{min(10, base*0.6):.1f}%",
        })
    pd.DataFrame(rows).to_csv(CSV_DST, index=False)
    print("  ✅ wc2026_predictions.csv  (generated from ELO)")

# ══════════════════════════════════════════════════════════════════
# STEP 9 — Verify everything
# ══════════════════════════════════════════════════════════════════
print("\n" + "="*55)
print("  VERIFICATION — Files in models/")
print("="*55)
total_mb = 0
for f in sorted(os.listdir(MODELS_DIR)):
    size = os.path.getsize(os.path.join(MODELS_DIR, f))
    total_mb += size
    print(f"  {'✅' if size > 0 else '❌'} {f:35s}  {size/1024:.1f} KB")
print(f"\n  Total size: {total_mb/1024/1024:.2f} MB")

print("\n" + "="*55)
print("  NEXT STEP — Push models/ folder to GitHub:")
print("="*55)
print("""
  Open a terminal in the same folder as this notebook, then run:

    git add models/
    git commit -m "Add trained model artifacts"
    git push

  Render will auto-redeploy in ~2 minutes.
  Then check: https://your-api.onrender.com/api/health
""")
print("🎉 All done!")
```
